```python
import keras
import numpy as np
import tensorflow
import matplotlib.pyplot as plt
import PIL 
```

```python
dataset = keras.utils.image_dataset_from_directory('../data/RSNA_AbTrauma/train_images/', batch_size=16, image_size=(256, 256))
```

```python
fig, axes = plt.subplots(4, 4)

for data, label in dataset:
    print(data.shape)
    print(label.shape)
    for i in range(data.shape[0]):
        for ax in axes.flat:
            image = data[i].numpy().astype('uint8')
            ax.imshow(image)
            ax.axis("off")
    break

plt.show()
```
